```python
X=df.drop('Churn',axis=1)
```

```python
y=df['Churn']
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, plot_confusion_matrix
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
scaler=StandardScaler()
```

```python
X_train_scaler=scaler.fit_transform(X_train)
```

```python
X_test_scaler=scaler.transform(X_test)
```

```python
def report_training(model,x_train,x_test,Y_train,Y_test):
    model.fit(x_train,Y_train)
    pred_train=model.predict(x_train)
    print(classification_report(Y_train,pred_train))
    plot_confusion_matrix(model,x_train,Y_train)
```

```python
def report_testing(model,x_train,x_test,Y_train,Y_test):
    model.fit(x_train,Y_train)
    pred_test=model.predict(x_test)
    print(classification_report(Y_test,pred_test))
    plot_confusion_matrix(model,x_test,Y_test)
```

```python
log_model=LogisticRegression()
```

```python
print('Logistic Regression on training data')
print('-----------------------------------------------')
report_training(log_model,X_train_scaler,X_test_scaler,y_train,y_test)
```

```python
print('Logistic Regression on testing data')
print('-----------------------------------------------')
report_testing(log_model,X_train_scaler,X_test_scaler,y_train,y_test)
```

```python
knn=KNeighborsClassifier()
```

```python
print('knn on training data')
print('-----------------------------------------------')
report_training(knn,X_train_scaler,X_test_scaler,y_train,y_test)
```

```python
print('knn on testing data')
print('-----------------------------------------------')
report_testing(knn,X_train_scaler,X_test_scaler,y_train,y_test)
```

```python
tree=DecisionTreeClassifier(max_depth=10, max_features=7)
```

```python
print('Decision tree on training data')
print('-----------------------------------------------')
report_training(tree,X_train,X_test,y_train,y_test)
```

```python
print('Decision tree on training data')
print('-----------------------------------------------')
report_testing(tree,X_train,X_test,y_train,y_test)
```

```python
tree.feature_importances_.argmax()
```

```python
rf_model=RandomForestClassifier(n_estimators=300,  random_state=101, max_depth=5, max_features=5)
```

```python
print('Random forest on training data')
print('-----------------------------------------------')
report_training(rf_model,X_train,X_test,y_train,y_test)
```

```python
print('Random forest on testing data')
print('-----------------------------------------------')
report_testing(rf_model,X_train,X_test,y_train,y_test)
```

```python
ada=AdaBoostClassifier(n_estimators=100)
```

```python
print('adaboost on training data')
print('-----------------------------------------------')
report_training(ada,X_train,X_test,y_train,y_test)
```

```python
print('Adaboost on testing data')
print('-----------------------------------------------')
report_testing(ada,X_train,X_test,y_train,y_test)
```

```python
gb_model=GradientBoostingClassifier(n_estimators=150)
```

```python
print('Gradient boosting on training data')
print('-----------------------------------------------')
report_training(gb_model,X_train,X_test,y_train,y_test)
```

```python
print('Gradient boosting on testing  data')
print('-----------------------------------------------')
report_testing(gb_model,X_train,X_test,y_train,y_test)
```
